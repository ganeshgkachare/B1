```python
def Engine_Name(x):
    Name = x.split(" ")[1]
    Engine_Name = Name.split("\xa0")[0]
    return Engine_Name
```

```python
def Engine_Model(x):
    Model = x.split(" ")[1]
    Engine_Model = Model.split("\xa0")[1]
    return Engine_Model
```

```python
def Equipment_Type(x):
    Equipment_Type=x.replace("Equipment Type ","")
    return Equipment_Type
```

```python
Q2['Engine Name'] = Q2['Engine'].apply(Engine_Name)
```

```python
Q2['Engine Model'] = Q2['Engine'].apply(Engine_Model)
```

```python
Q2['EquipmentType'] = Q2['EquipmentType'].apply(Equipment_Type)
```

```python
Q2.drop("Engine",axis=1,inplace=True)
```

```python
Output_Q2 = Q2[['Category', 'EquipmentMake', 'EquipmentModel', 'Equipment', 'Engine Name', 'Engine Model',
       'EquipmentType', 'Year', 'Part Number', 'Part Details 1',
       'Part Details 2', 'Filter Type', 'Quantity', 'Availability']]
```

```python
Output_Q2
```

```python
Output_Q2.to_excel(r"C:\Users\dell\Downloads\Python Test\Test 4\Output_Q2.xlsx", index=False)
```
